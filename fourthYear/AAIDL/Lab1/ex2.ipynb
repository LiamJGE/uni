```python
import numpy as np
# Single Instance Input
X = np.array(([0.4032, 0.4583]), dtype=float)
# Single Instance Output
y = np.array(([0.6200]), dtype=float)

# Class
class ANN(object):
    # constructor, set up weights
    def __init__(self):
        #parameters
        self.inputSize = 2
        self.outputSize = 1
        self.hiddenSize = 3
        #Weights (we are using two layers here), weights are random
        self.W1 = np.random.randn(self.inputSize, self.hiddenSize)
        self.W2 = np.random.randn(self.hiddenSize, self.outputSize)
    #sigmoid function
    def sigmoid(self, valuesIN):
        return 1/(1+np.exp(-valuesIN))
    #forward steps (remember slide 54)
    def forward(self, inputsX):
        # feedforward propagation
        self.Zi = np.dot(inputsX, self.W1) # Step 1
        self.Ai = self.sigmoid(self.Zi) # Step 2
        self.Z4 = np.dot(self.Ai, self.W2) # Step 3
        y_Hat = self.sigmoid(self.Z4) # Step 4
        return y_Hat
    def newWeights(self):
        # get new weights
        self.W1 = np.random.randn(self.inputSize, self.hiddenSize)
        self.W2 = np.random.randn(self.hiddenSize, self.outputSize)
```

```python
NN = ANN()
W1_Final = NN.W1
W2_Final = NN.W2
# Run 1 to get initial values, this could be the best weihghts
y_Hat = NN.forward(X)
loss = (0.5 * (np.square(y[0] - y_Hat[0])))
# back to sclae
print("Predicted Score:",round(y_Hat[0] * 100,2),"%")
print("Loss:", loss)
smallest_Loss = (0.5 * (np.square(y[0] - y_Hat[0])))
best_Prediction = round(y_Hat[0] * 100,2)
for item in range(2, 1001):
    # Update the weights to new random weights
    NN.newWeights()
    y_Hat = NN.forward(X)
    loss = (0.5 * (np.square(y[0] - y_Hat[0])))
    print("\n\nRun:", item)
    # back to scale
    print("Predicted Score:",round(y_Hat[0] * 100,2),"%")
    print("Loss:", loss)
    # If its the smallest loss, save the weights
    if loss < smallest_Loss:
        smallest_Loss = loss
        W1_Final = NN.W1
        W2_Final = NN.W2
        best_Prediction = round(y_Hat[0] * 100,2)
print("\n\n\n-----------------------------------")
print(" Best Model ")
print("-----------------------------------")
print("W1:", W1_Final)
print("\n\nW2:", W2_Final)
print("\n\nSmallest loss", smallest_Loss)
print("Best Prediction", best_Prediction,"%")


print("\n\nLoading best model...")

X2 = np.array(([0.5726, 0.5833],
               [0.4032, 0.4583],
               [0.6774, 0.6250],
               [0.6129, 0.5000]), dtype=float)

# Single Instance Output
y2 = np.array(([0.7500],
               [0.6200],
               [0.8000],
               [0.7800]), dtype=float)




nn2 = ANN()
nn2.W1 = W1_Final
nn2.W2 = W2_Final
y_Hat = nn2.forward(X2)
loss = (0.5 * (np.square(y2 - y_Hat)))
# back to sclae
print("Predicted Score:",y_Hat * 100,"%")
print("Loss:", loss) 
```
