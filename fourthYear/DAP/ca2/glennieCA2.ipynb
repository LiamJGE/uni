<h3>Import necessary packages</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import *
```

```python
data = pd.read_csv('Attrition_data.csv')
```

```python
for att in data.columns:
    print(att)
    print(data[att].describe())
    print()
```

```python
print("Data skewness")
print(data.skew())

print("\nData kurtosis")
print(data.kurt())
```

```python
plt.rc('font', size=10)
data.hist(figsize=(15,15))
plt.show()
```

```python
catData = data[data.columns[(data.dtypes=="object")]]
```

```python
catData
```

```python
for att in catData:
    sns.histplot(catData, x=att)
    plt.show()
```
