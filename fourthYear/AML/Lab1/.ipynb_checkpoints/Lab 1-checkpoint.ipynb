```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from statistics import *
```

```python
with open("weatherRaw.csv", "r", newline="") as file:
    wlist = list(csv.reader(file))

## Print in tabular way
for instance in wlist:
    for attr in instance:
        print("{:<15}".format(attr), end="")
    print()
```

```python
## Print number of instances
print("Number of instances " + str(len(wlist)))
```

```python
## Temperature min, max, avg
wlist = np.array(wlist)
temp = wlist[:,1].astype('float')
print("Temperature minimum " + str(min(temp)))
print("Temperature maximum " + str(max(temp)))
print("Temperature mean " + str(mean(temp)))
```

```python
## Humidity min, max, avg
hum = wlist[:,2].astype('float')
print("Humidity minimum " + str(min(hum)))
print("Humidity maximum " + str(max(hum)))
print("Humidity mean " + str(mean(hum)))
```

```python
## Class breakdown
add = 0
for el in wlist:
    if el[4] == "yes":
        add += 1    
print("Class breakdown\nYes: " + str(add) +"\nNo: " + str(len(wlist[:,1]) - add))
```

```python
## Temperature plot
plt.hist(temp)
oneStd = np.std(temp)
meanV = mean(temp)
plt.title("Checking for outliers")
plt.axvline(x=(meanV - 2*oneStd), label='Two STD', c="r")
plt.axvline(x=(meanV + 2*oneStd), c="r")
plt.axvline(x=(meanV - 3*oneStd), label='Three STD', c="g")
plt.axvline(x=(meanV + 3*oneStd), c="g")
plt.legend()
plt.show()
```

There are no clear outliers according to this histogram. 

```python
## Humidity plot
plt.hist(hum)
oneStd = np.std(hum)
meanV = mean(hum)
plt.title("Checking for outliers")
plt.axvline(x=(meanV - 2*oneStd), label='Two STD', c="r")
plt.axvline(x=(meanV + 2*oneStd), c="r")
plt.axvline(x=(meanV - 3*oneStd), label='Three STD', c="g")
plt.axvline(x=(meanV + 3*oneStd), c="g")
plt.legend()
plt.show()
```

There are no clear outliers according to this histogram. 

```python
with open("pima-indians-diabetes.csv", "r", newline="") as file:
    plist = list(csv.reader(file))

## Print in tabular way
for instance in plist:
    for attr in instance:
        print("{:<10}".format(attr), end="")
    print()
```

```python
## Print number of instances
print("Number of instances " + str(len(plist)))
```

```python
## Times pregnant min, max, avg
plist = np.array(plist)
preg = plist[:,0].astype(int)
print("Times pregnant minimum " + str(min(preg)))
print("Times pregnant maximum " + str(max(preg)))
print("Times pregnant mean " + str(mean(preg)))
```

```python
## Class breakdown
add = 0
for el in plist:
        add += int(el[8])    
print("Class breakdown\nYes: " + str(add) +"\nNo: " + str(len(plist[:,8]) - add))
```

```python
## Age plot
data = plist[:,7].astype('float')
plt.hist(data, align='left', edgecolor='black', linewidth=1.2)
oneStd = np.std(data)
meanV = mean(data)
print("Mean: " + str(meanV))
print("One StdV: " + str(oneStd))
plt.title("Age")
plt.axvline(x=(meanV - 2*oneStd), label='Two STD', c="r")
plt.axvline(x=(meanV + 2*oneStd), c="r")
plt.axvline(x=(meanV - 3*oneStd), label='Three STD', c="g")
plt.axvline(x=(meanV + 3*oneStd), c="g")
plt.legend()
plt.show()
```

Although there are many values outside of 2 standard variations, these values aren't really outliers because 
it is feasable that women over 55 years old would have taken part in the questionnaire or the test to collect
this data.

```python
## Skinfold thickness plot
data = plist[:,3].astype('float')
plt.hist(data, align='left', edgecolor='black', linewidth=1.2)
oneStd = np.std(data)
meanV = mean(data)
print("Mean: " + str(meanV))
print("One StdV: " + str(oneStd))
plt.title("Skinfold thickness")
plt.axvline(x=(meanV - 2*oneStd), label='Two STD', c="r")
plt.axvline(x=(meanV + 2*oneStd), c="r")
plt.axvline(x=(meanV - 3*oneStd), label='Three STD', c="g")
plt.axvline(x=(meanV + 3*oneStd), c="g")
plt.legend()
plt.show()
```

As we can see there are many instances where the skinfold is 0mm and this isn't normal. Therefore, this values must
be outliers despite not being outside of 2 standard deviations. 

```python
## BMI plot
data = plist[:,5].astype('float')
plt.hist(data, align='left', edgecolor='black', linewidth=1.2)
oneStd = np.std(data)
meanV = mean(data)
print("Mean: " + str(meanV))
print("One StdV: " + str(oneStd))
plt.title("Body Mass Index")
plt.axvline(x=(meanV - 2*oneStd), label='Two STD', c="r")
plt.axvline(x=(meanV + 2*oneStd), c="r")
plt.axvline(x=(meanV - 3*oneStd), label='Three STD', c="g")
plt.axvline(x=(meanV + 3*oneStd), c="g")
plt.legend()
plt.show()
```

There are some more clear outliers here because you can't have 0 BMI. It is very likely that these instances didn't have their BMI measured and a 0 was inserted instead. There are some unlikely BMI's outside of 2 standard deviations but they are possible so these instances should be taken into consideration.

```python
## Number of times pregnant plot
plt.hist(preg, align='left', edgecolor='black', linewidth=1.2)
oneStd = np.std(preg)
meanV = mean(preg)
print("Mean: " + str(meanV))
print("One StdV: " + str(oneStd))
plt.title("Number of times pregnant")
plt.axvline(x=(meanV - 2*oneStd), label='Two STD', c="r")
plt.axvline(x=(meanV + 2*oneStd), c="r")
plt.axvline(x=(meanV - 3*oneStd), label='Three STD', c="g")
plt.axvline(x=(meanV + 3*oneStd), c="g")
plt.legend()
plt.show()
```

Although the values to the right of 2 STDs are not very common nowadays, they are possible. Therefore, these instances should not be deleted from the dataset. It is likely that if we looked into the instances with a higher number of pregnancies, they would probably be associated with the older women that took part in the survey.
